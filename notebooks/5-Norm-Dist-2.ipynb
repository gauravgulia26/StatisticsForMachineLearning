```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = sns.load_dataset("titanic")
```

```python
def standardize_normal_variate(x: int, mu: int, sigma: int):
    return (x - mu) / sigma
```

```python
df["Standaridized_Age"] = df["age"].apply(
    standardize_normal_variate, args=(df["age"].mean(), df["age"].std())
)
```

```python
# plt.style.available
```

```python
plt.style.use("dark_background")
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

# Original Age distribution
df["age"].plot(kind="kde", label="Age", ax=axes[0], color="purple")
axes[0].set_title("KDE - Age")
axes[0].legend()
axes[0].set_xlabel("Age Values")

# Standardized Age distribution
df["Standaridized_Age"].plot(
    kind="kde", label="Standardized (Age)", ax=axes[1], color="white"
)
axes[1].set_title("KDE - Standard Normal Variate (Age)")
axes[1].legend(loc="upper right")
axes[1].set_xlabel("Standardized Age Values")

fig.suptitle(
    "Kernel Density Estimation: Age vs Standardized Age", fontsize=14, fontweight="bold"
)
plt.tight_layout(rect=[0, 0, 1, 0.95])

plt.tight_layout()
plt.show()
```
