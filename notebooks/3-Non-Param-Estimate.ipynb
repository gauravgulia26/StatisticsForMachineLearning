```python
from sklearn.neighbors import KernelDensity
from numpy.random import normal
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sample = normal(loc=10, scale=5, size=700)
sample_2 = normal(loc=30, scale=5, size=300)
```

```python
final_sample = np.hstack((sample, sample_2))
```

```python
model = KernelDensity(bandwidth=2)
```

```python
model.fit(final_sample.reshape(-1, 1))
```

```python
values = np.linspace(final_sample.min(), final_sample.max(), 200)
```

```python
prob = model.score_samples(values.reshape(-1, 1))
```

```python
prob_density = np.exp(prob)
```

```python
plt.hist(final_sample, bins=50, density=True)
plt.plot(values, prob_density)
plt.title("Non-Param Estimation")
plt.xlabel("Sample Values")
plt.ylabel("Prob Density")
plt.tight_layout()
plt.show()
```
