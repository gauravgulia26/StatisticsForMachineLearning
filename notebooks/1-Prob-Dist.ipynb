```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sample = pd.Series(np.linspace(1, 100, 1000))
```

```python
sample.plot(kind="kde")
plt.title("Prob Dist. (Continous Random Variable)")
plt.grid()
plt.xlabel("Marks Of Students")
plt.ylabel("Prob Density")
plt.tight_layout()
plt.show()
```

```python
a = np.random.randint(1, 100, 50)
```

```python
a = np.sort(a)
```

```python
discrete_sample = pd.Series(a)
```

```python
prob_dist = discrete_sample.value_counts().sort_index() / discrete_sample.shape[0]
```

```python
prob_dist.plot(kind="bar")
plt.title("Prob Dist. (Discrete Random Variable)")
plt.xlabel("Marks")
plt.ylabel("Probability")
plt.tight_layout()
plt.show()
```

```python
from scipy.stats import gaussian_kde

fig, axs = plt.subplots(1, 2, figsize=(12, 4))
# discrete probability distribution
prob_dist.plot(kind="bar", ax=axs[0])
axs[0].set_title("Prob Dist. (Discrete Random Variable)")
axs[0].set_xlabel("Marks")
axs[0].set_ylabel("Probability")
axs[0].grid(False)

# continuous probability distribution
sample.plot(kind="kde", ax=axs[1], color="lime")
axs[1].set_title("Prob Dist. (Continuous Random Variable)")
axs[1].set_xlabel("Marks")
axs[1].set_ylabel("Probability Density")

kde = gaussian_kde(sample)
x = np.linspace(min(sample), max(sample), 1000)
y = kde(x)
mask = (x >= 25) & (x <= 50)
axs[1].fill_between(
    x[mask], y[mask], color="orange", alpha=0.2, label="Area Under Graph 2550"
)

axs[1].legend()

fig.suptitle(
    "Comparison of Probability Distributions [ Continous, Discrete ] Random Variable"
)
plt.style.use("dark_background")
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```
