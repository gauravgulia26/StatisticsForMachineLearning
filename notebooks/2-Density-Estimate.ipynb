```python
import numpy as np
import pandas as pd
from numpy.random import normal
from scipy.stats import norm
import matplotlib.pyplot as plt
```

```python
mu, sigma = 50, 0.5
sample = normal(mu, sigma, 1000)
```

```python
plt.hist(sample, bins=20)
plt.title("Histogram of Sample")
plt.show()
```

```python
sample_mu, sample_sigma = sample.mean(), sample.std()
```

```python
dist = norm(sample_mu, sample_sigma)
```

```python
values = np.linspace(sample.min(), sample.max(), 100)
```

```python
prob_density = [dist.pdf(x) for x in values]
```

```python
plt.hist(sample, density=True, label="Original Distribution", color="skyblue")
plt.plot(values, prob_density, label="Estimated Density", linewidth=2, color="purple")
plt.legend()
plt.tight_layout()
plt.xlabel("Sample Values")
plt.ylabel("Prob Density")
plt.title("Parametric Estimation for N(mu,sigma)")
plt.style.use("dark_background")
plt.grid(axis="x")
plt.show()
```
